```python
import pandas as pd
import numpy as np
import multiprocessing
import seaborn as sns
import matplotlib.pyplot as plt
import collections

from time import time

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, recall_score, roc_auc_score, accuracy_score, precision_score, roc_curve, auc

from genetic_selection import GeneticSelectionCV

from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

```

## 1. Read data

```python
pd.options.display.max_columns = None

# read data
df = pd.read_csv('./data.csv')
print(df.shape)
df.describe()
```

```python
plt.pie([df['outcome'][df['outcome']==True].count(), df['outcome'][df['outcome']==False].count()],  autopct='%1.1f%%')
plt.show()

df.info()
```

## 2. Data Processing
1. drop columns which the missing ratio reach 70%
2. encode `gender` ('M':1, 'F':0) and `outcome`(True:1, False:0)
3. fill missing values


```python
# drop columns 
miss_ratio = np.sum(df.isna()) / len(df)
df = df.loc[:, miss_ratio < 0.7]

df['gender'] = df['gender'].apply(lambda x: 1 if x == 'M' else 0).astype(int)
df['outcome'] = df['outcome'].apply(lambda x: 1 if x is True else 0).astype(int)

# fit missing value
# imp = SimpleImputer(missing_values=np.nan, strategy='median')
imp = KNNImputer(n_neighbors=2, weights="uniform")

df = pd.DataFrame(imp.fit_transform(df), columns=df.columns)
df.to_csv('./data_process.csv', index=False)
```

```python
X_data, y_data = df.iloc[:, :-1], df.iloc[:, -1]


print(X_data.shape, y_data.shape)
print(y_data[y_data==0].count(), y_data[y_data==1].count())
print(y_data[y_data==0].count()/len(y_data), y_data[y_data==1].count()/len(y_data))
```

# Standardlization

```python
# scaler = MinMaxScaler()
scaler = StandardScaler()
X_data = pd.DataFrame(scaler.fit_transform(X_data), columns=X_data.columns)

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=False)
```

## Balance

4.1 SMOTE method

4.2 Oversampling

4.3 Undersampling


```python
class_ratio = 0.3

# 1. SMOTE method
sm = SMOTE(sampling_strategy=class_ratio)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
print('SMOTE: ',y_train_sm[y_train_sm==0].count(), y_train_sm[y_train_sm==1].count())

# 2. Oversampling
os = RandomOverSampler(sampling_strategy=class_ratio)
X_train_os, y_train_os = os.fit_resample(X_train, y_train)
print('Oversampling: ',y_train_os[y_train_os==0].count(), y_train_os[y_train_os==1].count())


# 3. Undersampling
us = RandomUnderSampler(sampling_strategy=0.6)
X_train_us, y_train_us = us.fit_resample(X_train, y_train)
print('Undersampling', y_train_us[y_train_us==0].count(), y_train_us[y_train_us==1].count())
```

## Feature Selection
3.1 Use DecisionTreeClassifier and Genetic Selection to select the features


```python
estimator = DecisionTreeClassifier(class_weight={1:100, 0:1})
selector = GeneticSelectionCV(estimator, cv=5, verbose=1, scoring="recall", n_population=20, crossover_proba=0.5,
                              mutation_proba=0.2, n_generations=50, crossover_independent_proba=0.5,
                              mutation_independent_proba=0.05, tournament_size=3, caching=True, max_features=25)

selector.fit(X_train_us, y_train_us)
print('Features:', X_data.columns[selector.support_])


estimator = DecisionTreeClassifier(class_weight={1:100, 0:1})
selector_sm = GeneticSelectionCV(estimator, cv=5, verbose=1, scoring="recall", n_population=20, crossover_proba=0.5,
                              mutation_proba=0.2, n_generations=50, crossover_independent_proba=0.5,
                              mutation_independent_proba=0.05, tournament_size=3, caching=True, max_features=25)

selector_sm.fit(X_train_sm, y_train_sm)
print('Features:', X_data.columns[selector_sm.support_])



# estimator = DecisionTreeClassifier(class_weight={1:100, 0:1})
# selector_us = GeneticSelectionCV(estimator, cv=5, verbose=1, scoring="recall", n_population=20, crossover_proba=0.5,
#                               mutation_proba=0.2, n_generations=50, crossover_independent_proba=0.5,
#                               mutation_independent_proba=0.05, tournament_size=3, caching=True, max_features=25)

# selector_us.fit(X_train_us, y_train_us)
# print('Features:', X_data.columns[selector_us.support_])
```

## 5. Model and Evaluation

```python
cur_X_train, cur_y_train = X_train_sm, y_train_sm
class_weight = {1:10, 0:1}



models = {
          'DT': DecisionTreeClassifier(class_weight=class_weight),
          'RF': RandomForestClassifier(class_weight=class_weight, n_estimators=200),
          'AdaBoost': AdaBoostClassifier(n_estimators=200)
}

models_params_grid = {
    'DT':{'criterion':['gini', 'entropy']},
    'RF':{'criterion':['gini', 'entropy']},
    'AdaBoost':{'learning_rate': [0.1, 0.2], 
                'estimator': [DecisionTreeClassifier(max_depth=4, class_weight = class_weight), 
                              DecisionTreeClassifier(max_depth=5, class_weight = class_weight)]}
}
# ,'SVM':{'kernel':['linear', 'poly', 'rbf'], 'C':[100, 10, 1.0, 0.1, 0.001]}

# Model training and Evaluation
result_dict = collections.defaultdict(list)
best_params = {}
clf_ls = {}

for name, model in models.items():
    print('----->', name)
    
    start_time = time()
    
    grid_search_clf = GridSearchCV(model, models_params_grid[name], scoring='recall', verbose=1, cv=3)
    grid_search_clf.fit(cur_X_train, cur_y_train)
    
    y_train_pred = grid_search_clf.predict(cur_X_train)
    y_train_proba = grid_search_clf.predict_proba(cur_X_train)
    y_test_pred = grid_search_clf.predict(X_test)
    y_test_proba = grid_search_clf.predict_proba(X_test)
    end_time = time()
    
    best_params[name] = grid_search_clf.best_params_
    clf_ls[name] = grid_search_clf
    
    result_dict['name'].append(name)
    result_dict['train_acc'].append(accuracy_score(cur_y_train, y_train_pred))
    result_dict['test_acc'].append(accuracy_score(y_test, y_test_pred))
    result_dict['train_auc'].append(roc_auc_score(cur_y_train, y_train_proba[:,1]))
    result_dict['test_auc'].append(roc_auc_score(y_test, y_test_proba[:,1]))
    result_dict['run_time'].append(end_time-start_time)
    print('-----> %s done in %0.4f s'%(name, end_time-start_time))
    print('-> train acc: %0.4f'%(result_dict['train_acc'][-1]))
    print('-> test acc: %0.4f'%(result_dict['test_acc'][-1]))
    print('-> ROC_AUC score on test: %0.4f'%(result_dict['test_auc'][-1]))
    print('->', grid_search_clf.best_params_)
    
    
    
# clf = RandomForestClassifier(n_estimators=100)
# clf.fit(X_train, y_train)
```
